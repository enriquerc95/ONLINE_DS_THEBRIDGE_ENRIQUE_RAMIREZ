```python
ciudades = ["LosAngeles", "SanDiego", "Sacramento", "Anaheim", "SantaAna"]
valores = [3884307, 1345895, 466488,336000,324528]
cities[["city", "population_total"]].sort_values(by="population_total" ,ascending=False)
city,population = cities["city"], cities["population_total"]
plt.figure(figsize=(8,8))
plt.bar(ciudades,valores, color=['blue', 'gray', 'olive', 'brown'], alpha = 0.5)
plt.xlabel('ciudades')
plt.ylabel('Habitantes(miles)')
plt.title('Ciudades ms pobladas de california');
plt.savefig("./imagen3.png")

```

```python
# 2. Histograma del rea total ocupada por las ciudades. (50 bins, fuente = "Bradley Hand itc", tamao 18, color azul acero)
font_dict= {
         "size" : 16,
         "family": "Bradley Hand itc", #tipo de letra
         "color": "black"
         }

area,num_ciudades = cities["area_total_km2"], cities["city"]
plt.figure(figsize=(8,8))
plt.hist(area, bins=50, color='steelblue')
plt.xlabel('Superficie($km^2$)', font_dict)
plt.ylabel('Nmero ciudades', font_dict)

plt.legend(["Hist Area"], loc = "upper right")
plt.savefig("./imagen3.png")

```

```python


fig3=plt.figure()
plt.scatter(cities.area_water_km2,cities.population_total/1000, c="g", alpha=0.3)
plt.xlabel(r"Area de agua en la ciudad ($km^2$)")
x=cities.area_water_km2
y=cities.population_total/1000
o1_x=cities.loc[cities["population_total"].idxmax(), "area_water_km2"]
o1_y=y.max()
o2_x=x.max()
o2_y= cities.loc[cities["area_water_km2"].idxmax(), "population_total"]/1000
x_text=(60)
y_text=(3500)

props_flecha={
    "arrowstyle":"->",
    "connectionstyle":"arc"
}
plt.title("Poblacin por kilmetro cuadrado de agua")
plt.annotate("Outliers", xy=(o1_x,o1_y), xytext=(x_text,y_text),arrowprops=props_flecha)
plt.annotate("Outliers", xy=(o2_x,o2_y), xytext=(x_text,y_text),arrowprops=props_flecha);
plt.savefig("./imagen3.png")
```
