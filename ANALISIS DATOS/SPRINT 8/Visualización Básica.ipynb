**Matplotlib fue originalmente escrito como una alternativa de Python para usuarios de MATLAB**, y gran parte de su sintaxis refleja ese hecho.
Las herramientas al estilo MATLAB están contenidas en la interfaz pyplot (``plt``).

#### Interfaz orientada a objetos

La interfaz orientada a objetos está disponible para situaciones más complicadas y cuando se desea tener más control sobre la figura.
En lugar de depender de alguna noción de una figura o ejes "activos", en la interfaz orientada a objetos las funciones de trazado son *métodos* de objetos ``Figure`` y ``Axes`` explícitos.
Para volver a crear el gráfico anterior usando este estilo de trazado, podrías hacer lo siguiente:


````python
# First create a grid of plots
# ax will be an array of two Axes objects
fig, ax = plt.subplots(2,2)
ax[0,0].plot(x, np.sin(x))
ax[1,1].plot(x, np.cos(x))
ax[1,1].plot(x, 0*x)
ax[0,1].plot(x, x)
#cuando hacemos la ceeacion del objeto figura, con el método axes, creamos el objeto figura y nos devuelve los subobjetos o subfiguras creadas.
#a x le pintamnos el seno de x y ese x nos da la posición 00
````

### Anatomía de una Figura y un Axe 

Para todos los gráficos de Matplotlib, comenzamos creando una figura y unos ejes (objeto tipo *axes*, ojo eje en inglés se dice *axis*).

````python
fig = plt.figure() #la variable fig, llamaremos al objeto figure aunque ahora solo hay una
ax = plt.axes()
#una figura puede ser uno o más axes, y lo que vemos son axes.
````

En Matplotlib, la *figura* (una instancia de la clase ``plt.Figure``) puede ser considerada como un solo **contenedor que contiene todos los objetos que representan ejes, gráficos, texto y etiquetas.**  

Los *ejes* (una instancia de la clase ``plt.Axes``) es lo que vemos arriba: un cuadro delimitador con **marcas y etiquetas, que eventualmente contendrá los elementos del gráfico que conforman nuestra visualización.**  

el axe, es la gráfica


### Primeros gráficos (axes)


Una vez que hemos creado unos ejes, podemos usar la función ``ax.plot`` para trazar algunos datos.

````python
fig = plt.figure()
ax = plt.axes()

x = np.linspace(0,10,100) #lista de valores x
ax.plot(x, np.sin(x)); #pintanos x y todos sus valores y la operacion de hacer np x, todos los valoresd que salen de ahí
#los ejes los decide automáticamente
````

### Ajustando el Gráfico: Colores y Estilos de Línea

El primer ajuste que podrías desear hacer a un gráfico es controlar los colores y estilos de línea.
La función ``plt.plot()`` toma argumentos adicionales que se pueden usar para especificar estos.
Para ajustar el color, puedes usar la palabra clave ``color``, que acepta un argumento de tipo string
El color se puede especificar de varias maneras:

````python
# specify color by name
plt.plot(x,np.sin(x-0), color = "blue")
````

````python
#como poner colores
plt.plot(x, np.sin(x - 1), color='g')           # short color code (rgbcmyk)
plt.plot(x, np.sin(x - 2), color='0.75')        # Grayscale between 0 and 1
plt.plot(x, np.sin(x - 3), color='#FFDD44')     # Hex code (RRGGBB from 00 to FF)
plt.plot(x, np.sin(x - 4), color=(1.0,0.2,0.3)) # RGB tuple, values 0 to 1
plt.plot(x, np.sin(x - 5), color='chartreuse'); # all HTML color names supported

#cambia la figura, porque cada vez que uso pyplot, hace una nueva.
````

Si no se especifica ningún color, Matplotlib automáticamente recorrerá un conjunto de colores predeterminados para múltiples líneas.

Del mismo modo, el estilo de línea se puede ajustar utilizando la palabra clave ``linestyle``:

````python
plt.plot(x, x+1, linestyle = "solid")
````

````python
plt.plot(x, x + 2, linestyle='dashdot')
plt.plot(x, x + 3, linestyle='dotted');

# For short, you can use the following codes:
plt.plot(x, x + 4, linestyle='-')  # solid
plt.plot(x, x + 5, linestyle='--') # dashed
plt.plot(x, x + 6, linestyle='-.') # dashdot
plt.plot(x, x + 7, linestyle=':');  # dotted
````

Si quisieras ser extremadamente conciso, estos códigos de ``linestyle`` y ``color`` pueden combinarse en un solo argumento no clave para la función ``plt.plot()``:

````python
plt.plot(x,x, "-g");
#linea sólida y verde
````

````python
plt.plot(x, x + 1, '--c') # dashed cyan
plt.plot(x, x + 2, '-.k') # dashdot black
plt.plot(x, x + 3, ':r');  # dotted red
````

Estos códigos de color de un solo carácter reflejan las abreviaturas estándar en los sistemas de color RGB (Rojo/Verde/Azul) y CMYK (Cian/Magenta/Amarillo/negro), comúnmente utilizados para gráficos de color digital.

Hay muchos otros argumentos clave que se pueden utilizar para ajustar finamente la apariencia del gráfico; para más detalles, como siempre, ir a la [documentación de plt.plot](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html#matplotlib.pyplot.plot)

Usarlo con objetos axe es "directo" (por resumir sólo en la versión concentrada):


````python
fig = plt.Figure()
ax = plt.axes()
ax.plot(x, x + 0, '-g')  # solid green
ax.plot(x, x + 1, '--c') # dashed cyan
ax.plot(x, x + 2, '-.k') # dashdot black
ax.plot(x, x + 3, ':r');  # dotted red
````

# Visualización con Matplotlib: Subplots

 Matplotlib tiene el concepto de *subplots*: grupos de ejes más pequeños que pueden existir juntos dentro de una sola figura.
Estos subplots podrían ser inserciones, cuadrículas de gráficos u otros diseños más complicados y además no tienen por qué ser del mismo tipo.  

````python
# %matplotlib inline
import matplotlib.pyplot as plt
plt.style.use("seaborn-v0_8-white")
import numpy as np
x = np.linspace(0,10,100)
````

### "A mano"

El método más básico para crear un eje es usar la función ``plt.axes``.
Como hemos visto anteriormente, de forma predeterminada esto crea un objeto de eje estándar que llena toda la figura.
``plt.axes`` también toma un argumento opcional que es una lista de cuatro números en el sistema de coordenadas de la figura.
Estos números representan ``[izquierda, inferior, ancho, altura]`` en el sistema de coordenadas de la figura, que varía de 0 en la esquina inferior izquierda de la figura a 1 en la esquina superior derecha de la figura.

Por ejemplo, podríamos crear un eje secundario en la esquina superior derecha de otro eje estableciendo la posición *x* y *y* en 0.65 (es decir, comenzando en el 65% del ancho y el 65% de la altura de la figura) y las extensiones *x* y *y* en 0.2 (es decir, el tamaño del eje es el 20% del ancho y el 20% de la altura de la figura):


````python
ax1 = plt.axes() #crea figura con parámetros por defecto
ax2 = plt.axes([0.5, 0.5,0.2,0.3])
#nos crea en el centro de la figura, la esquina inferior izquierda del nuevo eje.
````

````python
#cambiando la figura
ax1 = plt.axes() #crea figura con parámetros por defecto
ax2 = plt.axes([0.1, 0.1,0.3,0.3]) #0.2 y 0.3, indica el porcentaje
#nos crea en el centro de la figura, la esquina inferior izquierda del nuevo eje.
````

````python
#cambiando la figura
ax1 = plt.axes() #crea figura con parámetros por defecto
ax2 = plt.axes([0.1, 0.1,0.3,0.3]) #0.2 y 0.3, indica el porcentaje
ax1.plot(x, 2*x)
ax2.plot(x,np.sin(x)); #podemos también poner un plot
#nos crea en el centro de la figura, la esquina inferior izquierda del nuevo eje.
````

El equivalente de este comando dentro de la interfaz orientada a objetos es ``fig.add_axes()``. Usemos esto para crear dos ejes apilados verticalmente:

````python
#crear dos ejes, 1 encima del otro.
fig = plt.figure()
ax1 = fig.add_axes([0.1, 1, 1, 2],
                   xticklabels=[], ylim=(-1.2, 1.2)) #crear en el 10% del margen del todo
ax2 = fig.add_axes([0.1, 0.1, 1.5, 1],
                   ylim=(-1.2, 1.2))

x = np.linspace(0, 10)
ax1.plot(np.sin(x))
ax2.plot(np.cos(x));

#no se suele utilizar
#crea el primero, se come el 10% de margen en los dos. Podemos superponer. Pero es más complicada de ver.
````

### Con cuadrículas sencillas

Las columnas o filas alineadas de *subplots* o gráficas formando cuadrículas o matrices son una necesidad lo suficientemente común como para que Matplotlib tenga varias rutinas de conveniencia que las hacen fáciles de crear.

El nivel más bajo de éstas es ``plt.subplot()``, este comando tiene tres argumentos enteros: el número de filas, el número de columnas y el índice de la gráfica que se creará en este esquema, que va desde la esquina superior izquierda hasta la esquina inferior derecha.


matriz de subplots o gráficas de 3x5, y las numeraremos con su índice:

````python
plt.figure(figsize = (15, 10)) #tamaño de la figura
for i in range(1,16): #matriz de 15 elementos
    plt.subplot(3,5,i) #subplot, sin s, la i empieza a contar en 1 y no en 0
    plt.text(0.5, 0.5, str((3,5,i)), fontsize = 16, ha = "center") #con text, añadimops un tecxto a una figura. 
             #Fontsize fuente, ha alineación
#crea matriz de 3x5 y al decir la y, sabe donde asignarrlo.
````

El comando ``plt.subplots_adjust`` se puede usar para ajustar el espaciado entre estos gráficos.
El siguiente código utiliza el comando orientado a objetos equivalente, ``fig.add_subplot()``:

````python
#podemos alinear el ancho entre subplots
plt.figure() #tamaño de la figura
fig.subplots_adjust(hspace = 0.4, wspace = 0.8)  #margen verticcal h, margen horizontal w
for i in range(1,7): #matriz de 15 elementos
    plt.subplot(2,3,i) #subplot, sin s, la i empieza a contar en 1 y no en 0
    plt.text(0.5, 0.5, str((2,3,i)), fontsize = 16, ha = "center") #con text, añadimops un tecxto a una figura. 
             #Fontsize fuente, ha alineación
#crea matriz de 3x5 y al decir la y, sabe donde asignarrlo.
````

### La cuadrícula completa de una vez

**La aproximación anterior puede volverse bastante tediosa si queremos crear una cuadrícula o matriz grande, especialmente si se desea ocultar las etiquetas del eje x y y en las tramas internas.**  

Para este propósito, ``plt.subplots()`` es la herramienta más fácil de usar (nota la ``s`` al final de ``subplots``). En lugar de crear una sola subtrama, esta función crea una cuadrícula completa de subtramas en una sola línea, devolviéndolas en un array de NumPy.
Los argumentos son el número de filas y el número de columnas, junto con palabras clave opcionales ``sharex`` y ``sharey``, que le permiten especificar las relaciones entre diferentes ejes.  


Aquí crearemos una cuadrícula de $2 \times 3$ subtramas, donde todos los ejes en la misma fila comparten su escala en el eje y, y todos los ejes en la misma columna comparten su escala en el eje x:

````python
#Es el mejor, usamos subplots, que nos crea un objeto figura y un array de objetos axe, con las dimensiones elegidas, en el que cada elemento delo array es una subgráfica con las dimensiones elkegidas, (3x2 3 filas de dos columnas, por ejemplo.
fig,ax = plt.subplots(2,3, sharex = "col", sharey = "row") #nos da la cuadrículas compartierndo el eje x y el y . #en ax tengo un array con los subgráficos numerados.
````

````python
# Los axes están en un array bidimensional accedido por [fila,columna] tipo python, empezando en cero
for i in range(2):
    for j in range(3):
        ax[i,j].text(0.5,0.5,str((i,j)), fontsize = 18, ha = "center")
fig
#nos hace la misma matriz, pero en cada una nos pone el texto centrado en la que la cadena es cada una de sus cordenadas y con acceder a ax y esas cordenadas, pordemos cambiar el gráfico.
````

Si queremos acceder a cualquiera de los gráficos podemos acceder a partir del elemento del array de objetos en "ax" (tal como hemos añadido los textos):

````python
ax[1,1].plot(x,np.sin(x))
fig #nos añade el seno donde queríamos, al añadir el eje, desplaza los textos.
````

````python
# Te dejo aquí como haríamos para quitar el texto:
fig, ax = plt.subplots(2, 3, sharex = "col", sharey = "row")
textos = np.full(fill_value= None, shape= ax.shape) #crea nuevo array
for i in range(2):
    for j in range(3):  
        textos[i,j] = ax[i, j].text(0.5, 0.5, str((i, j)),
                      fontsize=18, ha='center')
textos[0,2].remove()
textos[1,1].remove()
ax[1,1].plot(x,np.sin(x));
#cuanndo añadimos el elemento text, crea otro onbjeto que tenemos que almacenar en algún sitio para referenciarlo
#se selecciona de uno a uno.
````
